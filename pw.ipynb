np.array():
This function converts input data (like lists or tuples) into a NumPy array and always creates a copy of the input data by default, regardless of whether the input is already an ndarray. This means any changes made to the original data after creating the array will not affect the array created by np.array().

np.asarray():
This function also converts input data into a NumPy array but does not create a copy if the input is already an ndarray. It is more memory efficient because it avoids unnecessary copying. Changes to the original ndarray will be reflected in the array created by np.asarray()

np.asanyarray():
This function is similar to np.asarray() but allows subclasses of ndarray (like matrices) to pass through unchanged. If the input is an ndarray subclass, it returns that subclass; otherwise, it converts the input to an ndarray.



Shallow Copy:
A shallow copy creates a new object but does not create copies of nested objects within it. Instead, it references the same nested objects as the original. Changes made to mutable nested objects in the shallow copy will reflect in the original object.

Deep Copy:
A deep copy creates a new object and recursively adds copies of nested objects found in the original. Changes made to nested objects in a deep copy do not affect the original object.

```python
import numpy as np

# Step 1: Create a 5x6 array with random integers between 1 and 10
random_array = np.random.randint(1, 11, size=(5, 6))

# Output the generated array
print("Random Array:")
print(random_array)

# Step 2a: Extract all even integers from the array
even_integers = random_array[random_array % 2 == 0]

# Output the even integers
print("\nEven Integers:")
print(even_integers)

# Step 2b: Extract all odd integers from the array
odd_integers = random_array[random_array % 2 != 0]

# Output the odd integers
print("\nOdd Integers:")
print(odd_integers)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample time-series dataset
data = {
    'Date': pd.date_range(start='2024-10-01', periods=10, freq='D'),
    'Temperature': [22, 21, 23, 24, 25, 26, 24, 22, 23, 24],
    'Humidity': [55, 60, 58, 57, 56, 59, 62, 61, 60, 58]
}

df = pd.DataFrame(data)

# Plot Temperature and Humidity with two y-axes
fig, ax1 = plt.subplots()

# Plotting Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature', color='r')
ax1.tick_params('y', colors='r')

# Create a second y-axis for Humidity
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity', color='b')
ax2.tick_params('y', colors='b')

# Set the title of the plot
plt.title('Temperature and Humidity Over Time')

# Display the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create NumPy array with 1000 samples from a normal distribution
data = np.random.randn(1000)

# Plot histogram with 30 bins
plt.hist(data, bins=30, density=True, alpha=0.6, color='g', label='Histogram')

# Overlay the PDF of a normal distribution
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, 0, 1)
plt.plot(x, p, 'k', linewidth=2, label='Normal PDF')

# Label the axes
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')

# Set the title of the plot
plt.title('Histogram with PDF Overlay')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create two random arrays
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrants based on x and y values
def get_quadrant(x, y):
    if x >= 0 and y >= 0:
        return 'First Quadrant'
    elif x < 0 and y >= 0:
        return 'Second Quadrant'
    elif x < 0 and y < 0:
        return 'Third Quadrant'
    else:
        return 'Fourth Quadrant'

# Apply the function to create a color category for each point
quadrants = [get_quadrant(x_val, y_val) for x_val, y_val in zip(x, y)]

# Create a DataFrame for plotting
df = pd.DataFrame({'x': x, 'y': y, 'Quadrant': quadrants})

# Create a Seaborn scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='x', y='y', hue='Quadrant', palette='Set1')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Show the legend and plot
plt.legend(title='Quadrant')
plt.show()
```

```python
from bokeh.plotting import figure, show, output_notebook
import numpy as np

# Prepare the output for Jupyter Notebook
output_notebook()

# Create a figure
p = figure(title="Sine Wave Function", x_axis_label='X-axis', y_axis_label='Y-axis')

# Generate x and y values for the sine wave
x = np.linspace(0, 4 * np.pi, 100)
y = np.sin(x)

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Sine Wave", line_width=2)

# Add grid lines (enabled by default in Bokeh, but reinforced)
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'

# Show the plot
show(p)
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np

# Prepare the output for Jupyter Notebook
output_notebook()

# Randomly generated categorical data and values
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create a ColumnDataSource for the data
source = ColumnDataSource(data=dict(categories=categories, values=values, color=np.random.choice(['#e41a1c', '#377eb8', '#4daf4a', '#984ea3'], len(categories))))

# Create a figure for the bar chart
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           x_axis_label='Categories', y_axis_label='Values')

# Add bars with color and tooltips
p.vbar(x='categories', top='values', width=0.9, color='color', source=source)

# Add hover tool to display exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Show the plot
show(p)
```
